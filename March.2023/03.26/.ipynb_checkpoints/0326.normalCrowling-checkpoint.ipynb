```python
import pandas as pd
import bs4, requests
```

```python
targetUrl = "https://sparkkorea.com/%ed%80%b4%ec%a6%88/"
```

```python
try:
    resq = requests.get(targetUrl)
except Exception as e:
    print(e)
```

```python
resq
```

```python
webPageSrc = resq.text
```

```python
webPageSrc
```

```python
targetObj = bs4.BeautifulSoup(webPageSrc, "html.parser")
```

```python
targetObj
```

```python
divs = targetObj.find(name = "div", attrs = {"id":"id_spark_quiz"})
```

```python
divs
```

```python
aTags = divs.findAll(name = "a")
```

```python
aTags
```

```python
quizName = aTags[0].text
```

```python
quizLink = aTags[0].attrs["href"]
```

```python
nameList = []
```

```python
linkList = []
```

```python
lenAtags = len(aTags)
```

```python
for i in range(0, lenAtags):
    quizName = aTags[i].text
    quizLink = aTags[i].attrs["href"]
    
    nameList.append(quizName)
    linkList.append(quizLink)
```

```python
pD01 = pd.DataFrame(zip(nameList, linkList), columns = ["퀴즈이름", "퀴즈링크"])
```

```python
rowList = []
```

```python
for j in range(0, lenAtags):
    rowName = aTags[i].text, aTags[i].attrs["href"]
    rowList.append(rowName)
```

```python
pD02 = pd.DataFrame(rowList, columns = ["퀴즈이름", "퀴즈링크"])
```
