```python
# calendar 함수를 호출
import calendar

# 상수값 CONSTANT
END_OF_LAST_MONTH = 31
END_OF_MONTH = 12

def dayCalc(_inputData, _pastData):
    """
    8자리 형태로 현재 연월일 정보를 입력받아 회귀일수 만큼 이전의 연월일 형태로 정보를 출력하는 함수입니다.
    결과값을 문자열 형태로 출력합니다.
    _pastData : 8자리 형태의 문자열 데이터 형식으로 출력됩니다.
    사용변수:
    _inputData : year+month+days 정보로 구성된 8자리 연월일 데이터 입력을 요합니다. 정수와 문자열의 타입으로 입력이 가능합니다. ex)20230423 ex2)"20230423"
    _pastData : _inputData로부터 회귀하고자 하는 일자 데이터입니다. 정수와 문자열 타입으로 입력이 가능합니다. ex)190 ex2)"190"
    본 함수 사용을 위해서는 calendar 함수가 필요합니다.
    return : _result(str)로 'YYYYMMDD' 형태로 함수 결과값을 출력합니다.
    """
    
    
    # 01. 변수의 설정
    
    # _inputData의 [0:4] 인덱스를  _yearData 변수에 저장. 인덱스 사용을 위해 정수 데이터를 문자열 형태로 변환하고, 이후 출력한 인덱스 값을 정수 데이터로 변환합니다.
    _yearData = int(str(_inputData)[0:4])
    
    # _inputData의 [4:6] 인덱스를  _monthData 변수에 저장. 상동.    
    _monthData = int(str(_inputData)[4:6])

    # _inputData의 [6:] 인덱스를  _dateData 변수에 저장. 상동.
    _dateData = int(str(_inputData)[6:])

    # _pastData의 치환
    _pastData = (int)(_pastData)
    
    #_target01 변수에 _dateData - _pastData 값을 저장합니다.
    _target01 = _dateData - _pastData


    # 02. 연산
    
    
    # 02 - 1. 현재 일자가 목표 회귀 일자보다 큰 경우의 if 연산
    
    if(_target01 > 0):
        # zfill(2)를 사용하여 만약 각 문자열 데이터가 두 자리 미만이라면 앞 자리에 0을 채워넣습니다.
        # 현재 일자 - 목표 회귀 일자 데이터를 사용하여 일자 처리를 마감합니다.
        _result = str(_yearData) + str(_monthData).zfill(2) + str(_target01).zfill(2)

    # 02-2. 현재 일자와 회귀 희망 일자가 동일한 경우의 연산
    
    elif(_target01 == 0):
        
        # 현재 월자 데이터에서 1을 감합니다.
        _monthData = _monthData - 1
        
        # 만약 1을 감하였을 때, monthData가 0이 되어버렸다면
        if(_monthData == 0):
           
            # yearData에서 1을 차감하고,
            _yearData = _yearData - 1
           
            # monthData에 12를 더해줍니다.(년도를 1년 앞으로 당기고 기준 월 데이터를 12월로 세팅)
            _monthData = _monthData + END_OF_MONTH
        
        # 위 케이스에 해당하지 않는 경우, 아래 연산으로 직행합니다.
        else:
            pass
        
        # target02 변수에 calender.monthrange(_yearData, _monthData)[1] 값을 저장합니다.
        # calendar.monthrange(year, month) 기능은 해당 연도 대상 월의 총 일자를 구하여 튜플 형식으로 데이터를 출력하는 기능입니다.
        # 해당 기능을 활용할 경우, 대상 월의 마지막 일자 데이터를 확보할 수 있습니다.
        # 이에 기존 월자 정보에서 1을 차감하고, 해당 월자의 마지막 일자를 세팅하여 정보를 출력합니다.
        _target02 = calendar.monthrange(_yearData, _monthData)[1]
        
        # 결과값 출력 구문. 이하 case 2-1과 동일.
        _result = str(_yearData) + str(_monthData).zfill(2) + str(_target02).zfill(2)
    
    
    
    # 2-3. 현재 일자보다 희망 회귀 일자 데이터가 더 큰 경우의 연산
    
    else:
        
        #_remainDays 변수에 두 일자 간의 차이 값을 저장합니다.
        _remainDays = _pastData - _dateData
        
        # 현재 월자 데이터에서 1을 차감한 값을 월자 데이터에 저장합니다.
        _monthData = _monthData - 1

        # 만약 위 월자 연산을 통해 월자 데이터가 0이 되었다면 연차 데이터에서 1을 차감하고 월자 데이터에 12를 더합니다.        
        if(_monthData == 0):
            _yearData = _yearData - 1
            _monthData = _monthData + 12

        
        # 만약 _remainDays 변수의 값이 31 이상이라면, 아래의 연산을 반복합니다.
        while(_remainDays >= END_OF_LAST_MONTH):
            
            # _remainDays 에서 calendarcalendar.monthrange(현재 연차, 현재 월자)[1] 연산 값을 차감합니다.
            _remainDays = _remainDays - calendar.monthrange(_yearData, _monthData)[1]
            
            # 차감 후 월자 데이터에서 1을 차감한 값을 월자 데이터에 저장합니다.
            _monthData = _monthData - 1
            
            # 해당 연산 중에도 월자 데이터가 0이 된다면 연차 데이터에서 1을 차감하고 월자에 12를 더하는 연산을 수행합니다.
            if(_monthData == 0):
                _yearData = _yearData - 1
                _monthData = _monthData + END_OF_MONTH   
        
        
        # 위의 while 연산을 통해 _)remainDays 변수 값이 31 이하로 떨어진다면, 아래의 연산을 수행합니다.
        else:
            _target03 = calendar.monthrange(_yearData, _monthData)[1] - _remainDays
            # _target03에 해당 연, 월의 총 일자 - _remainDays를 저장합니다.
            
            if(_target03 == 0):
            # 만약 _target03이 0이라면
                _monthData = _monthData - 1
                _target03 = calendar.monthrange(_yearData, _monthData)[1]
                # _monthData에서 1을 감하고, 이를 적용한 연월의 총 일자를 _target03에 저장합니다.
                if(_monthData == 0):
                    _yearData = _yearData - 1
                    _monthData = _monthData + END_OF_MONTH
                    _target03 = calendar.monthrange(_yearData, _monthData)[1]
                # 해당 연산 간 _monthData가 -이 된다면 _yearData에서 1을 감하고, _monthData에 12를 더한 후 연월 총 일자를 _target03에 저장합니다.
                    
            elif(_target03 < 0):
            # _target03이 음수 값을 지니는 경우(2월 등의 특수 케이스)
                    _monthData = _monthData - 1
                    _target03 = calendar.monthrange(_yearData, _monthData)[1]+_target03
                    # 대상 _monthData에서 1을 감하고, 이를 적용한 연월 총 일자에서 음수인 _target03을 더한 값을 _target03 값으로 저장합니다.
                    if(_monthData == 0):
                        _yearData = _yearData - 1
                        _monthData = _monthData + END_OF_MONTH
                        _target03 = calendar.monthrange(_yearData, _monthData)[1]+_target03 
                    # 연산 간 _monthData가 0이 된 경우 위와 동일하게 _yearData와 _monthData의 조작 후, 연산합니다.
            else:
                pass
        
        # 이하 출력 구문. 상동.
        _result = str(_yearData) + str(_monthData).zfill(2) + str(_target03).zfill(2)

    # 결과값으로 _result 변수 값을 출력합니다.
    return _result
```
