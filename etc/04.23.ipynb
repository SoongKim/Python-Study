```python
import numpy as np
```

```python
customerData = pd.read_csv("../../../dataset/customerdata.csv")
```

```python
customerData.info()
```

```python
customerData["AGE_NEW"] = \
np.where(customerData.PRODUCTAGE <= 1, 1,\
         np.where(customerData.PRODUCTAGE <= 2, 2, 3))
```

```python
customerData.head()
```

```python
condition = customerData.PRODUCTAGE <= 1
condition2 = (customerData.PRODUCTAGE > 1) & \
(customerData.PRODUCTAGE <= 2)
condition3 = customerData.PRODUCTAGE > 2
```

```python
customerData.loc[condition, "NEW_AGE"] = 1
customerData.loc[condition2, "NEW_AGE"] = 2
customerData.loc[condition3, "NEW_AGE"] = 3
```

```python
customerData.head()
```

```python
customerData
```

```python
customerData["product_age_new"] = \
np.where(customerData.PRODUCTAGE < 1, 1, \
         np.where(customerData.PRODUCTAGE < 2, 2, \
                  np.where(customerData.PRODUCTAGE < 3, 3, 5)))
```
