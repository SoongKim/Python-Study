```python
import pandas as pd
import bs4, requests
# pandas와 bs4, requests 모델 호출
```

```python
targetUrl = "https://www.ssg.com/search.ssg?target=all&query=65%EC%9D%B8%EC%B9%98%20tv"
# 목표 Url. 이번에는 ssg 닷컴에서 65인치 tv를 검색한 결과창을 대상으로 하였다.
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'
resp = requests.get(targetUrl,headers = {'User-Agent':user_agent} )
# 우회 준비
```

```python
webSrc = resp.text
# targeturl로부터 확보한 데이터 중 text 형식 분량을  webSrc라는 변수로 지정
```

```python
targetObj = bs4.BeautifulSoup(webSrc, "html.parser")
# bs4를 통해 정제한 webSrc를 targetObj로 지정
```

```python
targetObj
```

```python
###################################
```

```python
titleTags = targetObj.findAll(name = "div", attrs = {"class":"ssg_price_ko show_gl hide_ko"})
# 해당 이름의 태그에 부합하는 자료를 추출
```

```python
titleTags[0].text
# 가격 정보가 출력되었음을 확인
```

```python
len(titleTags)
# 해당 태그 정보의 갯수를 확인
```

```python
lenPrice = len(titleTags)
# 확인한 태그 정보 갯수를 변수로 지정
```

```python
priceList = []
# 빈 리스트를 생성
```

```python
for i in range(0, lenPrice):
    priceValue = titleTags[i].text
    priceList.append(priceValue)
# for문을 통해 빈 리스트에 가격 정보를 삽입. 이를 가격 정보 리스트로 활용
```

```python
len(priceList)
# 총 114개의 가격 정보가 priceList에 편성되었다.
```

```python
##################################
```

```python
nameTags = targetObj.findAll(name = "em", attrs = {"class":"tx_en"})
# 해당 태그에 해당하는 데이터를 추출
```

```python
nameTags
```

```python
len(nameTags)
# 가격정보 114개보다 3개 많음을 확인. 그중 최후미 세 개가 상품명이 아닌, 안내 문구임을 파악
```

```python
lenNameTags = len(nameTags)
# nameTags의 갯수를 lenNameTags라는 변수로 지정
```

```python
nameList = []
# 빈 리스트를 생성
```

```python
for j in range(0, lenNameTags):
    nameValue = nameTags[j].text
    nameList.append(nameValue)
# for문을 통해 빈 리스트에 상품명 정보를 삽입. 추후 이를 상품명 리스트로 활용
```

```python
nameList
```

```python
# nameList 마지막 세 컨텐츠는 사이트 내 안내 문구를 담고 있으므로 제거.
# 이를 제거할 시 가격 리스트와 동일한 114개 컨텐츠를 지닌다.
```

```python
len(nameList)
```

```python
del nameList[len(nameList)-1]

del nameList[len(nameList)-1]

del nameList[len(nameList)-1]
```

```python
tvDf = pd.DataFrame(priceList, nameList)
# 가격 리스트와 상품명 리스트를 pandas를 통해 DataFrame화 하였다.
```

```python
tvDf.columns = ["65인치 TV 상품명/가격"]
# column 명으로 위 명칭을 지정하였다.
```

```python
tvDf
# 사이트를 크롤링하고, 이로부터 확보한 데이터를 바탕으로 만들어낸 데이터프레임형 자료 데이터.
```

```python
tvDf.to_csv("././tvDf.csv", encoding = "ms949")
# csv 파일 하나는 ms949로 인코딩하여 저장하였다.
```
