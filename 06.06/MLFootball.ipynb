```python
for i in range(0, targetData.shape[0]):
    if("Goalkeeper" in targetData.position[i]):
        targetData.position[i] = "Goalkeeper"
    elif("midfield" in targetData.position[i]):
        targetData.position[i] = "midfield"
    elif("Defender" in targetData.position[i]):
        targetData.position[i] = "defender"
    elif("Winger" in targetData.position[i]):
        targetData.position[i] = "Winger"
    else:
        targetData.position[i] = "Attacker"
```

```python
# 포지션 데이터를 골키퍼, 수비수, 미드필더, 윙어, 공격수로 구분하였다.
```

```python
tvcData = targetData.club.value_counts()
```

```python
clubList = list(tvcData.index)
```

```python
clubData = list(0 for i in range(0, len(clubList)))
```

```python
for i in range(0, len(clubList)):
    clubData[i] = targetData.loc[targetData.club == clubList[i]]
```
