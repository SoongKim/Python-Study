```python
inData = pd.concat( [indata[0], indata[1]], axis = 0 )
```

```python
# pd.concat 기능을 통해 두 데이터를 합하였습니다.
```

```python
columns = [inData.columns]
```

```python
columns
```

```python
inData = inData.rename(columns=lambda x: re.sub(r'^(\d+)(.+)', r'\2\1', x))
```

```python
# 컬럼명 가장 앞에 존재하는 숫자들을 뒤로 이전. Regular Expression 사용
```

```python
inData.columns = [re.compile(r'[^\w\s]').sub('', col) for col in inData.columns]
```

```python
# 컬럼명에서 특수문자를 제거
```

```python
inData.columns
```

```python
inData.columns = inData.columns.str.replace(' ', '_')
```

```python
inData.columns
```

```python
# 여기까지의 작업을 통해 inData의 컬럼에서 숫자를 뒤로 보내고,
# 특수문자를 제거하고, 띄어쓰기를 _로 치환하였습니다.
```

```python
inData.바닥슬라브두께
```

```python
inData = inData.dropna(subset=['바닥슬라브두께'])
```

```python
# 'NaN'이라고 적혀있는 문자열 데이터를 제거
```

```python
import numpy as np
```

```python
inData.loc[inData['바닥슬라브두께'] == '-', '바닥슬라브두께'] = None
inData = inData.dropna(subset=['바닥슬라브두께'])
```

```python
# 기타 이상치 제거
```

```python
inData['바닥슬라브두께'] = inData['바닥슬라브두께'].astype(float)
```

```python
# 이후 바닥슬라브두께 컬럼의 데이터를 실수 타입으로 전환
```

```python
for col in inData.columns:
    inData[col] = inData[col].apply(lambda x: pd.to_numeric(x, errors='ignore'))
```
