### 이동집계함수

```python
### 이동집계함수 : 이동평균, 이동 표준편차 등 구간을 이동하면서 특정 통꼐지표를 산출하는 방식
### 5인 경우 나를 기준으로 왼쪽(2) 오른쪽(2)에 날개를 펼침 
```

```python
# 데이터프레임명["컬럼명"].rolling(window = 5 , center =True, min_periods = 1).mean()
```

```python
# window = 5 : 구간설정
# center =True: 기준점, true로 설정 안하면 내가 맨 마지막에 있다
#min_periods : 최소구간
#              5인 경우 나를 기준으로 왼쪽(2) 오른쪽(2)에 날개를 펼침,양옆에 값이 2개 없으면 
#              셋팅 안하면 중간이 아닐경우 null값을 넣게 됨
```

```python
selloutData = pd.read_csv("../../stdcode6/dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutData
```

```python
# 1. 정렬하기
# 정렬은 기본키로 해야함!
```

```python
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
```

```python
sortedData=selloutData.sort_values(by = sortKey)
```

```python
sortedData
```

```python
# QTY를 기준으로 rolling 할 것
```

```python
sortedData["QTY"].rolling(window = 5 , center = True).mean()
```

```python
sortedData["MA5"] = sortedData["QTY"].rolling(window = 5 , center = True).mean()
```

```python
sortedData
```

```python
sortedData["MA5"] = sortedData["QTY"].rolling(window = 5 , center = True, min_periods=2).mean()
```

```python
sortedData
```

```python
row01 = ["A01", "ST0001", 0.3]
row02 = ["A02", "ST0002", 0.5]
row03 = ["A03", "ST0003", 0.9]
```

```python
promoData = pd.DataFrame([row01, row02, row03], columns = ["REGIONID", "PRODUCTNAME", "PROMORATE"])
```

```python
promoData
```

```python
pd.merge(left=joinSellout,
          right=promoData,
          left_on=["REGIONID", "PRODUCT"],
          right_on=["REGIONID", "PRODUCTNAME"],
        how="left")
```

```python
# 고객이 요청한 데이터는 무조건 왼쪽으로
```

```python
pd.merge(left=joinSellout,
          right=promoData,
          left_on=["REGIONID", "PRODUCT"],
          right_on=["REGIONID", "PRODUCTNAME"],
        how="right")
```

```python
pd.merge(left=joinSellout,
          right=promoData,
          left_on=["REGIONID", "PRODUCT"],
          right_on=["REGIONID", "PRODUCTNAME"],
        how="inner")
```

```python
pd.merge(right=joinSellout,
          left=promoData,
          right_on=["REGIONID", "PRODUCT"],
          left_on=["REGIONID", "PRODUCTNAME"],
        how="left")
```

```python
pd.merge(right=joinSellout,
          left=promoData,
          right_on=["REGIONID", "PRODUCT"],
          left_on=["REGIONID", "PRODUCTNAME"],
        how="right")
```

```python
pd.merge(right=joinSellout,
          left=promoData,
          right_on=["REGIONID", "PRODUCT"],
          left_on=["REGIONID", "PRODUCTNAME"],
        how="inner")
```
