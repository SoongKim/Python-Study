```python
selloutData = pd.read_csv("../../../stdcode6/dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutData
```

```python
sortKey = ["REGIONID", "PRODUCT", "YEARWEEK"]
```

```python
sortedData = selloutData.sort_values(sortKey).reset_index(drop=True)
```

```python
sortedData["MA5"] = sortedData["QTY"].rolling(window=5, center=True, min_periods = 3).mean()
```

```python
sortedData.MA5
```

```python
sortedData["SEASONALITY"] = sortedData["QTY"] / sortedData["MA5"]
```

```python
sortedData.SEASONALITY
```

```python
SEASONALITY_1 = sortedData.loc[(sortedData.SEASONALITY >= 1)].reset_index(drop=True)
```

```python
SEASONALITY_1.loc[:, ["REGIONID", "PRODUCT", "YEARWEEK", "QTY", "SEASONALITY"]]
```
