```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
orclID = 'kopo'
orclPW = 'kopo'
orclIP = '192.168.110.112'
orclPort = '1521'
orclDbName = 'orcl'

engine = create_engine('oracle+cx_oracle://{}:{}@{}:{}/{}'.format(orclID, orclPW, orclIP, orclPort, orclDbName))
```

```python
joinData1 = pd.read_sql_query('SELECT * FROM JOINDATA1', engine)
```

```python
joinData2 = pd.read_sql_query('SELECT * FROM JOINDATA2', engine)
```

```python
joinData1 = joinData1.drop([0])
```

```python
joinData1.reset_index(drop=True)
```

```python
joinData2
```

```python
joinTotal = pd.merge(left=joinData1,
                     right=joinData2,
                     left_on=["mata_tin"],
                     right_on=["mata_tin"],
                     how="left")
```

```python
sortKey = ["mata_tin"]
```

```python
joinTotal.sort_values(sortKey).reset_index(drop=True)
```
