```python
import pandas as pd
```

```python
import bs4, requests
```

```python
targetUrl = "https://sparkkorea.com/%ED%85%8C%EC%8A%A4%ED%8A%B8/"
```

```python
resp = requests.get(targetUrl)
```

```python
webSrc = resp.text
```

```python
htmObj = bs4.BeautifulSoup(webSrc, "html.parser")
```

```python
htmObj
```

```python
tableTag = htmObj.find(name="table", attrs={"id":"test_table"})
```

```python
tableTag
```

```python
tBody = tableTag.find(name="tbody")
```

```python
tBody
```

```python
trs = tBody.findAll(name="tr")
```

```python
columnList = []

rowList = []
```

```python
eachRow = trs[0]
```

```python
trs[0].findAll("td")
```

```python
tds = eachRow.findAll("td")
```

```python
for j in range(0, len(trs)):
    
    eachRow = trs[j]
    tds = eachRow.findAll("td")
    for i in range(0, len(tds)):
        c1Data = tds[i].text
        columnList.append(c1Data)
    rowList.append(columnList)

    columnList = []
```

```python
Df01 = pd.DataFrame(rowList)
```
