```python
import pandas as pd
```

```python
import bs4, requests
```

```python
targetUrl = ("https://sparkkorea.com/%ed%85%8c%ec%8a%a4%ed%8a%b8/")
```

```python
try:
    resp = requests.get(targetUrl)
except Exception as e:
    print(e)
```

```python
webPageSrc = resp.text
```

```python
targetObj = bs4.BeautifulSoup(webPageSrc, "html.parser")
```

```python
targetObj
```

```python
tableTag = targetObj.find(name = "table", attrs={"id":"test_table"})
```

```python
tableTag
```

```python
tbody = tableTag.find("tbody")
```

```python
tbody
```

```python
trs = tbody.findAll("tr")
```

```python
trs
```

```python
columnList = []
```

```python
rowList = []
```

```python
eachRow = trs[0]
```

```python
eachRow
```

```python
tds = eachRow.findAll("td")
```

```python
for i in range(0, len(trs)):
    eachRow = trs[i]
    tds = eachRow.findAll("td")
    for i in range(0, len(tds)):
        c1Data = tds[i].text
        columnList.append(c1Data)
    rowList.append(columnList)
        
    columnList = []
```
