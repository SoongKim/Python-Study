```python
# np.where 방식은 빠르다!
# 그러나 복잡한 로직을 적용하기에는 힘들다!
```

```python
## 함수를 적용할 수 있다.
```

```python
def productAgeDefine(inValue):
    #inValue = 1.5
    outValue = 1

    if inValue < 1:
       outValue = 1
    elif inValue < 2:
       outValue = 2
    elif inValue < 3:
       outValue = 3
    else:
       outValue = 4
    return outValue
```

```python
customerData["PROAGE_NEW"] = customerData.PRODUCTAGE.apply(productAgeDefine)
```
