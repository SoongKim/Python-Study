```python
# 평균 : mean(), 개수 : count(), 최대값 : max(), 최소값 : min()
# 표준편차 : stdev() 정도를 주로 사용한다.
```

count : 아이템의 전체 개수
first, last : 첫 번째, 마지막 아이템
mean : 평균값
median : 중간값
min, max : 최소값, 최대값
std, var : 표준편차와 분산
prod : 그룹 데이터의 곱
sum : 그룹 데이터의 합
quantile : 사분위수

```python
custData = pd.read_csv("../../../../stdcode6/dataset/customerdata.csv")
```

```python
groupKey01 = ["EMI"]
```

```python
custData.groupby(groupKey01)[["AVGPRICE"]].mean()
```

```python
custData.groupby(groupKey01)[["AVGPRICE"]].median()
```

```python
custData.groupby(groupKey01)[["AVGPRICE"]].quantile()
```

```python
custData.groupby(groupKey01)[["AVGPRICE"]].agg(["mean", "median", "max", "min", "var", "std", "count"])
```

```python
custData.groupby(groupKey01)[["AVGPRICE"]].agg(["max"])
```

```python
custData.groupby(groupKey01)[["AVGPRICE"]].max()
```

```python
custData.groupby(groupKey01).std()
```

```python
groupData = custData.groupby(groupKey01)[["AVGPRICE"]].std()
```

```python
groupData.reset_index()
```
