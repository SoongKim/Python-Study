```python
import pandas as pd
```

```python
selloutData = pd.read_csv("../../stdcode6/dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutData.head()
```

```python
sortKeys = ["REGIONID", "PRODUCT", "YEARWEEK"]
```

```python
sortedData = selloutData.sort_values(sortKeys)
```

```python
sortedData["MA5"] = sortedData["QTY"].rolling(window=5, center=True, min_periods=1).mean()
```

```python
sortedData["SEASONALITY"] = sortedData["QTY"] / sortedData["MA5"]
```

```python
sortedData.head()
```

```python
sortedData["YEAR"] = sortedData.YEARWEEK.astype(str).str[0:4]
sortedData["WEEK"] = sortedData.YEARWEEK.astype(str).str[4:]
```

```python
sortedData.head()
```

```python
groupKeys = ["REGIONID", "PRODUCT", "WEEK"]
```

```python
sortedData.groupby(groupKeys)["SEASONALITY"].agg(["mean"]).reset_index()
```

```python
# 지역ID, 제품명, 주차로 groupby 하였으므로 각 지역 / 상품 / 주차 별 SEASONALITY가 출력된다.
```
