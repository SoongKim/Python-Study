```python
import pandas as pd
import bs4, requests
```

```python
targetUrl = "https://www.ssg.com/search.ssg?target=all&query=%EC%9E%84%EC%97%B0%EC%88%98%20%EC%86%8D%EC%B4%88"
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'
resp = requests.get(targetUrl,headers = {'User-Agent':user_agent} )
```

```python
targetPage = resp.text
```

```python
targetObj = bs4.BeautifulSoup(targetPage, "html.parser")
```

```python
targetObj
```

```python
nameTag = targetObj.findAll(name = "em", attrs = {"class":"tx_ko"})
```

```python
nameTag[0].text
```

```python
nameTag
```

```python
del nameTag[0::2]
```

```python
nameTag
```

```python
nameList = []
```

```python
for i in range(0, len(nameTag)):
    nameValue = nameTag[i].text
    nameList.append(nameValue)
```

```python
nameList
```

```python
len(nameList)
```

```python
priceTags = targetObj.findAll(name = "div", attrs = {"class":"ssg_price_ko show_gl hide_ko"})
```

```python
len(priceTags)
```

```python
priceList = []
```

```python
for j in range(0, len(priceTags)):
    priceValue = priceTags[j].text
    priceList.append(priceValue)
```

```python
priceList
```

```python
import pandas as pd
```

```python
fishDf = pd.DataFrame(priceList, nameList)
```

```python
fishDf.columns = ["상품명/가격"]
```

```python
fishDf
```

```python
fishDf.to_csv("./Python-Study/fishDf.csv", encoding = "ms949")
```
