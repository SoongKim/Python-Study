```python
import pandas as pd
import numpy as np
```

```python
customerData = pd.read_csv("../../../Python_Script/dataset/customerdata.csv")
```

```python
customerData
```

```python
def productAge(productAge):
    
    if productAge < 1 :
        return 1
    elif productAge < 2 :
        return 2
    elif productAge < 3 :
        return 3
    else:
        return 4
```

```python
customerData["product_new_age"] = customerData.PRODUCTAGE.apply(productAge)
```
