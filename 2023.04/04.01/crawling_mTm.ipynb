```python
import pandas as pd
import requests, bs4
```

```python
targetUrl = "https://www.ssg.com/search.ssg?target=all&query=%EC%B2%AD%EB%B0%94%EC%A7%80"
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'
resp = requests.get(targetUrl,headers = {'User-Agent':user_agent} )
```

```python
targetHtml = resp.text
```

```python
targetObj = bs4.BeautifulSoup(targetHtml, "html.parser")
```

```python
targetObj
```

```python
kokrTags = targetObj.findAll(name = "em", attrs={"class":"tx_ko"})
```
