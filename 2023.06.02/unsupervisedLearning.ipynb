#### 2. 모델 적용

```python
kmeanModel = KMeans(random_state=11, n_clusters=3)
```

```python
features = ["SW", "DB", "AND"]
```

```python
stDataTarget = stData.loc[:, features]
```

#### 자율학습예측

```python
predictCluster = kmeanModel.fit_predict(stDataTarget)
```

```python
stData["CLUSTER"] = predictCluster
```

```python
predictCluster
```
