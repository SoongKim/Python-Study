```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
ynMap = {"Y":1, "N":0}
```

```python
# 딕셔너리 타입은 중괄호{}로 선언한다.
# 딕셔너리 타입의 구조 : {KEY : VALUE}로 구성되어 있다.
```

```python
targetData.PROMOTION.map(ynMap)
```

```python
targetData["NEW_HO_YN"] = targetData.HOLIDAY.map(ynMap)
```

```python
targetData.head()
```

```python
# np.where보다 map 함수를 사용하면 훨신 간편하게 데이터를 정돈할 수 있다.
```

```python
targetData["HO_YN"] = \
np.where(targetData["HOLIDAY"] == "Y", 1, 0)
```

```python
targetData["PRO_YN"] = \
np.where(targetData["PROMOTION"] == "Y", 1, 0)
```
